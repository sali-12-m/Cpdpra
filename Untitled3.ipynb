```python
import pandas as pd
```

```python
df = pd.read_csv("/content/StudentPerformanceFactors.csv")

print(df.shape)
print(df.columns)
print(df.head())

print(df.isnull().sum())
```

**Insight 1**

Study hours alone do not fully determine exam performance; motivation level significantly influences how effectively those hours translate into results. Students with high motivation consistently achieve higher exam scores even when their study hours are similar to those with low motivation, indicating that the quality of study is more important than quantity. Additionally, while students from higher-income families and private schools tend to perform better on average, this advantage diminishes among highly motivated students, suggesting that motivation can partially compensate for socioeconomic and institutional differences.

```python
df = df.drop_duplicates()

df = df.ffill()
```

```python
corr = df.corr(numeric_only=True)['Exam_Score'].sort_values(ascending=False)
print(corr)
```

**Insight 2** — Key Drivers of Exam Score

Previous academic performance shows the strongest correlation with exam scores (r = X), indicating that students’ past performance is a strong predictor of future outcomes. Hours studied and attendance also show moderate positive correlations, suggesting that consistent effort contributes to improved results. However, weaker correlations for factors like sleep indicate that not all lifestyle variables have a direct measurable impact.

```python
cross = pd.crosstab(
    df['Gender'],
    df['Motivation_Level'],
    values=df['Exam_Score'],
    aggfunc='mean'
)
print(cross)
```

**Insight 3** — Multi-Dimensional Analysis

Although slight fluctuations are observed across Gender and Motivation_Level, Exam_Score remains highly stable across all groups, with only minimal variation (≈66.7 to 67.8).

Both male and female students show nearly identical performance patterns across motivation levels, indicating that motivation does not meaningfully change academic outcomes in this dataset. Additionally, there is no visible interaction effect between gender and motivation, as trends remain parallel and consistent rather than diverging.

Overall, the data suggests that Exam_Score is largely independent of both Gender and Motivation_Level, implying that other unmeasured factors are likely driving performance differences.

```python
cohort = df.groupby(['Family_Income', 'School_Type'])['Exam_Score'].mean()
print(cohort)
```

**Insight 4** — Cohort Analysis

Exam performance shows only marginal variation across Family_Income and School_Type combinations, with all cohort averages tightly clustered around 66.8 to 67.9. While students from higher-income families exhibit a very slight performance advantage overall, this difference is small and not practically significant.

Across school types, no consistent advantage for either private or public institutions is observed. In fact, performance differences between school types fluctuate across income groups, indicating no stable or systematic school effect.

Importantly, the expected moderating effect—where school type amplifies or reduces the impact of family income—is not strongly supported by the data, as income-related differences remain minimal within both private and public schools.

```python
top = df[df['Exam_Score'] >= df['Exam_Score'].quantile(0.9)]
bottom = df[df['Exam_Score'] <= df['Exam_Score'].quantile(0.1)]

print(top.describe())
print(bottom.describe())
```

**Insight 5** — Percentile Analysis

The comparison between the top 10% and bottom 10% of students reveals a clear and substantial performance separation driven by behavioral and academic engagement factors rather than innate ability alone.

Top-performing students demonstrate significantly higher Hours_Studied (≈25.5 vs 14.9) and Attendance (≈91.5% vs 67.9%), indicating that they invest considerably more time in structured learning and maintain stronger classroom consistency. They also show higher Previous_Scores, suggesting a cumulative advantage built over time.

In contrast, bottom-performing students exhibit lower study time, reduced attendance, and fewer tutoring sessions, reinforcing the pattern that lack of consistency in learning behaviors strongly correlates with lower academic outcomes.

Importantly, sleep and physical activity remain relatively similar across both groups, suggesting that these factors are not primary differentiators of performance in this dataset.

```python
df['Study_Efficiency'] = df['Exam_Score'] / df['Hours_Studied']

eff = df.groupby('Motivation_Level')['Study_Efficiency'].mean()
print(eff)
```

**Insight 6** — Derived Metric

The derived metric Study_Efficiency (Exam_Score per Hour Studied) reveals only slight variation across Motivation_Level groups, with values ranging narrowly from approximately 3.76 to 3.92.

Students with high motivation show marginally higher efficiency compared to medium and low motivation groups, indicating they achieve slightly more academic output per hour of study. However, the difference is very small in magnitude, suggesting that motivation has only a weak association with study efficiency in this dataset.

Medium and low motivation groups perform almost identically, reinforcing the idea that efficiency is relatively stable across motivation categories rather than strongly stratified by them.
